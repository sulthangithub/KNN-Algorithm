```python
#data visulation Matplotlib use
#hist in histogram (matplotlib)
import matplotlib.pyplot as plt
for label in (tel_data.columns):
    plt.hist(tel_data[tel_data['class']=='g'][label],color='red', label='gamma' ,alpha= 0.5, density=True)
    plt.hist(tel_data[tel_data['class']=='h'][label],color='green' ,label='hadron', alpha= 0.5, density= True)
    plt.title(label)
    plt.ylabel("Probability")
    plt.xlabel(label)
    plt.legend()
    plt.show()
```
